```r
library(dplyr)
library(ggplot2)
```

```r
df <- read.csv('League of Legends 2021 World Championship Play-In Groups Statistics - Raw Data.csv')
head(df)
```

```r
colnames(df)
```

```r
df$kda <- round((df$Kills + df$Assists) / df$Deaths, digits=2)
head(df)
```

```r
df %>%
group_by(kda) %>%
arrange(desc(kda))
```

```r
Top <- df %>%
filter(Position == 'Top') 

Top
```

```r
TopSorted <- Top %>%
count(Champion) %>%
arrange(desc(n))

TopSorted
```

```r
ggplot(df, aes(x=Champion, y=kda)) + geom_bar(stat='identity') + 
     theme(axis.text.x = element_text(angle = 90, hjust = 1), legend.position = "None")
```

```r
ggplot(TopSorted, aes(x=Champion,y=n )) + geom_point()
```

```r
p1 <- ggplot(data=Top, aes(x=Champion)) + geom_bar(aes(fill=Player), position='dodge') +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
p1
```

```r
p1 + facet_wrap(~Team) + coord_flip() + theme_minimal()
```

```r
ggplot(Top, aes(x=Champion, y=Kills)) + geom_point(aes(color=Player)) + facet_wrap(~Player) +
    coord_flip() + theme_bw()
```
